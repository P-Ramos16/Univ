```python
import numpy as np
import matplotlib.pyplot as plt


# Tempo inicial e final
ti = 0
tf = 100

# Massa
m = 1
k = 1
xxEqui = 1.5
Em = 0.75

# Posio inicial
xx0 = np.sqrt(np.sqrt(2 * Em / k) + xxEqui**2)

# Velocidade inicial
vx0 = 0

# Acelerao inicial
ax0 = 0

# dt incremento do tempo e n numero de intervalos
dt = 0.001
n = int((tf - ti) / dt)

Epot = np.empty(n + 1)
Emec = np.empty(n + 1)

# Vetor tempo (n+1 para garantir que nao falta o ultimo dado (Ex: t[10]))
t = np.linspace(ti, tf, n + 1)

# Vetor velocidade (empty e no zeros para no alterar 
# muito o resultado se faltar analisar um dado)
xx = np.empty(n + 1)
vx = np.empty(n + 1)
ax = np.empty(n + 1)

# Introduzir x0 e v0 nos vetores da posio e velocidade
xx[0] = xx0
vx[0] = vx0
ax[0] = ax0

# Preencher os vetores x, v, a
for i in range(n):
    ax[i] = -2 * k * (xx[i]**2 - xxEqui**2) * xx[i] / m
    vx[i + 1] = vx[i] + ax[i] * dt
    xx[i + 1] = xx[i] + vx[i+1] * dt
    
    
    Epot[i] = (0.5 * k * (xx[i]**2 - xxEqui**2)**2)

    Emec[i] = 0.5 * m * vx[i]**2 + Epot[i]
    

# Plot
plt.plot(t, xx, label="x",  linestyle='-', linewidth=3, color="turquoise")
plt.xlabel("x (m)")
plt.ylabel("Emec (J)")
plt.legend()         # Legenda s aparece com isto
plt.grid()
plt.show()

plt.plot(xx, Epot, label="Emec",  linestyle='-', linewidth=3, color="mediumspringgreen")
plt.xlabel("x (m)")
plt.ylabel("Emec (J)")
plt.legend()         # Legenda s aparece com isto
plt.grid()
plt.show()

plt.plot(xx, Emec, label="Emec",  linestyle='-', linewidth=3, color="mediumspringgreen")
plt.xlabel("x (m)")
plt.ylabel("Emec (J)")
plt.legend()         # Legenda s aparece com isto
plt.grid()
plt.show()

    
    
```
