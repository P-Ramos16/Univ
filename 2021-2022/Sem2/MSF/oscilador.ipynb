```python
import matplotlib.pyplot as plt
import numpy as np
import math


dt=0.01
tf=300.0
t0=0

n=np.int((tf-t0)/dt+0.1)
print('\nn : ', n)

x = np.zeros(n+1)
vx = np.zeros(n+1)
ax = np.zeros(n+1)
ay = np.zeros(n+1)
y = np.zeros(n+1)
vy = np.zeros(n+1)
t = np.zeros(n+1)

vx[0] = 0
x[0] = 4
massa = 1
kapa =1

b = 0.01
f0 = 2
wf = 10
for i in range(n):
    t[i+1]=t[i]+dt
    ax[i]=-kapa*x[i]/massa + (-b*vx[i] + f0 * math.cos(wf*t[i]))/massa
    vx[i+1]=vx[i]+ax[i]*dt
    x[i+1]=x[i]+vx[i+1]*dt
plt.grid()
plt.plot(t,x,'-g')
plt.plot(t, vx,"--b")
plt.plot(t,np.zeros(n+1),"-k")
plt.legend(["Distancia(s)","Velocidade(m/s)"])
```
