```python
import numpy as np
import matplotlib.pyplot as plt

# Carro
xZeroC = 0
vZeroC = 70 * 1000 / 3600
acelC = 0

# Policia
xZeroP = 0
vZeroP = 0
acelP = 2

t = np.linspace(0, 100)


xC = lambda t : (xZeroC + vZeroC * t + 0.5 * acelC * t ** 2)

xP = lambda t : (xZeroP + vZeroP * t + 0.5 * acelP * t ** 2)


fig, ax = plt.subplots()


plt.plot(t, xC(t), linestyle='-.', linewidth=3, color="red", markersize=12)

plt.plot(t, xP(t),  linestyle='-.', linewidth=3, color="blue", markersize=12)


plt.xlabel("t (s)")
plt.ylabel("x (m)")

ax.set(xlim=(0, 30.01), xticks=np.arange(0, 30.01, step=5),
    ylim=(0, 500.01), yticks=np.arange(0, 500.01, step=50))

plt.show()
```
