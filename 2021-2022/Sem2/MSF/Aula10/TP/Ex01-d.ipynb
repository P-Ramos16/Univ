```python
#   OSCILADORES: f = -k x(t)

import numpy as np
import matplotlib.pyplot as plt

# Gravidade
g = 9.8

# Massa
m = 1

# Tempo inicial e final
ti = 0
tf = 300

# Area, r, u, Cres e Potencia
K = 1
w0 = np.sqrt(K / m)
b = 0.01
F0 = 2
w = 10

# Posio inicial
xx0 = 4

# Velocidade inicial
vx0 = 0

# dt incremento do tempo e n numero de intervalos
dt = 0.001
n = int((tf - ti) / dt)

# Vetor tempo (n+1 para garantir que nao falta o ultimo dado (Ex: t[10]))
t = np.linspace(ti, tf, n + 1)

# Vetor velocidade (empty e no zeros para no alterar 
# muito o resultado se faltar analisar um dado)
xx = np.empty(n + 1)
vx = np.empty(n + 1)
ax = np.empty(n + 1)

Emec = np.empty(n + 1)

# Introduzir x0 e v0 nos vetores da posio e velocidade
xx[0] = xx0
vx[0] = vx0


# Preencher os vetores x, v, a
for i in range(n):    
    
    ax[i] = (-K * xx[i] / m) + ( -b * vx[i]) + (F0 * np.cos(w * t[i]))
    vx[i + 1] = vx[i] + ax[i] * dt
    xx[i + 1] = xx[i] + vx[i + 1] * dt
    
  

# Encontrar os dados do paraquedista quando ele chega ao solo (x = 0)
for i in range(n):
    if ((0 - dt) < vx[i] < (0 + dt)):
        print("T =", t[i],":")
        print("t >  | xx[t] >")
        print(t[i],"  " , xx[i])
        print("")
        plt.plot(t[i], xx[i], "o", markersize="10", linewidth=5, color="red")
        break

for i in range(n):
    if ((0 - dt) < xx[i] < (0 + dt)):
        print("T =", t[i],":")
        print("t >  | xx[t] >")
        print(t[i],"  " , xx[i])
        print("")
        plt.plot(t[i], xx[i], "o", markersize="10", linewidth=5, color="red")

# Plot
plt.plot(t, xx, label="Posio", linestyle='-', linewidth=3, color="turquoise")
plt.plot(t, t * 0, linestyle='-', linewidth=1, color="black")
plt.xlabel("t (s)")
plt.ylabel("v (m/s)")
plt.legend()         # Legenda s aparece com isto
plt.grid()
plt.show()
```
