```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.80
tZero = 0
tFinal = 5
xZero = 0
vZero = 0
tDelta = 0.01
cl = "blue"

fig, ax = plt.subplots()

def func():
    N = int((tFinal - tZero) / tDelta + 0.1)

    print(N)

    t = np.zeros(N+1)
    x = np.zeros(N+1)
    v = np.zeros(N+1)



    v[0] = vZero
    x[0] = xZero
    t[0] = tZero



    for i in range(N):
        t[i + 1] = t[i] + tDelta
        v[i] = g * t[i]
        x[i + 1] = x[i] + v[i] * tDelta



    plt.plot(t, x, linestyle='-.', linewidth=3, color=cl, markersize=12)



    
    



func()
cl = "red"
tDelta = 0.5
func()



plt.xlabel("t (s)")
plt.ylabel("x (m)")

ax.set(xlim=(0, 5.01), xticks=np.arange(0, 5.01, step=0.5),
    ylim=(0, 120.01), yticks=np.arange(0, 120.01, step=10))

plt.show()
```
