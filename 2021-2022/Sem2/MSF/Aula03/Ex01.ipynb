```python
import numpy as np
import matplotlib.pyplot as plt


tZero = 0
xZero = 0
vZero = 0
tDelta = 0.1
N = 20

t = np.linspace(0, N * tDelta)

vX = lambda y : (9,81 * y)

xX = lambda t: (9.81 * t * t * tDelta)




fig, ax = plt.subplots()


plt.plot(t, xX(t), linestyle='-.', linewidth=3, color="springgreen", marker='+', markersize=12)

plt.xlabel("t (s)")
plt.ylabel("x (m)")

ax.set(xlim=(0, 2.01), xticks=np.arange(0, 2.01, step=0.25),
    ylim=(0, 5.01), yticks=np.arange(0, 5.01, step=0.5))

plt.show()


```
