```python
import numpy as np
import matplotlib.pyplot as plt


# Variáveis práticas

L = np.array([92, 113, 133, 153, 171.5, 194, 207.5, 222])

X = np.array([1.0, 1.2, 1.4, 1.6, 1.8, 2, 2.2, 2.3])


# plot
# modelo das variáveis práticas
plt.plot(L, X, label="f(x)", linestyle='-.', linewidth=2.0, color="blue", marker='+', markersize=12, markerfacecolor='blue')
plt.ylabel("v(m/s)")
plt.xlabel("t(s)")
plt.legend()         # Legenda só aparece com isto
plt.grid()
plt.show()
```
