```python
import numpy as np
import matplotlib.pyplot as plt



# Variáveis práticas

L = np.array([0.6950, 4.363, 15.53, 38.74, 75.08, 125.2, 257.9, 344.1, 557.4, 690.7])

X = np.array([200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100])

# plot
fig, ax = plt.subplots()

plt.xlabel("X (Joules)")
plt.ylabel("L (Kelvin)")

 # modelo das variáveis práticas
ax.plot(L, X, linestyle='-.', linewidth=2.0, color="blue", marker='+', markersize=12, markerfacecolor='blue')

ax.set(xlim=(0, 700), xticks=np.arange(0, 700.1, step=100),
       ylim=(0, 1200), yticks=np.arange(0, 1200.1, step=100))

plt.show()
print("A relação entre o tempo e a distância percorrida não é linear")
```
