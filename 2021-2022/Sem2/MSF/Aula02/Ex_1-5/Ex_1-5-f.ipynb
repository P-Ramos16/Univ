```python
import numpy as np
import matplotlib.pyplot as plt



def calcRegLin(L, X):
    # Calculo dos somatórios

    xySum = np.sum(L * X)

    xSum = np.sum(L)

    ySum = np.sum(X)

    x2Sum = np.sum(L ** 2)

    y2Sum = np.sum(X ** 2)



    # Calculo r2 (ao quadrado)

    N = np.size(L)

    m = np.round(( N * xySum - (xSum * ySum) ) / ( N * x2Sum - (xSum ** 2) ), 8)

    b = np.round( ( x2Sum * ySum - xSum * xySum ) / ( N * x2Sum - (xSum ** 2) ) , 8)

    r2 = np.round( (N * xySum - xSum * ySum) ** 2 / ( ( N * x2Sum - xSum ** 2 ) * ( N * y2Sum - ySum ** 2 ) ), 8)

    mDelta = np.round( abs(m) * np.sqrt(((1/r2) - 1 )/(N - 2)), 8)

    bDelta = np.round( mDelta * np.sqrt(x2Sum/N), 8)
      
    
    print("     r² >", r2)
        
    # plot

    fig, ax = plt.subplots()

        # reta calculada
    plt.plot(range(0, 250), range(0, 250)*m+b)

    plt.xlabel("X (s)")
    plt.ylabel("L (cm)")
        
    ax.set(xlim=(0, 250), xticks=np.arange(0, 251, step=25),
        ylim=(0, 2.5), yticks=np.arange(0, 2.6, step=0.5))

    
    return
    
    


# Variáveis práticas

L   =  np.array([92, 113, 133, 153, 171.5, 194, 207.5, 222])

LAlt = np.array([92, 113, 133, 153, 250, 194, 207.5, 222])

#                                   ↑↑↑  Valor alterado (171.5 --> 250)

X = np.array([1.0, 1.2, 1.4, 1.6, 1.8, 2, 2.2, 2.3])



calcRegLin(L, X)
calcRegLin(LAlt, X)

plt.show()

print(" > O coeficiente de determinação afasta-se de 1 \n quando os valores afastam-se da reta encontrada \n pelos valores médios")
```
