```python
import numpy as np
import matplotlib.pyplot as plt

# Tempo inicial e final
ti = 0
tf = 100

# Massa
m = 1.5
k = 1.2
alpha = -0.01
xxEqui = 0

Em = 2

# Posio inicial
xx0 = np.sqrt(np.sqrt(2 * Em / k)**2 + xxEqui**2)

# Velocidade inicial
vx0 = 0

# Acelerao inicial
ax0 = 0

# dt incremento do tempo e n numero de intervalos
dt = 0.01
n = int((tf - ti) / dt)


# Vetor tempo (n+1 para garantir que nao falta o ultimo dado (Ex: t[10]))
t = np.linspace(ti, tf, n + 1)

# Vetor velocidade (empty e no zeros para no alterar 
# muito o resultado se faltar analisar um dado)
Epot = np.empty(n + 1)
Emec = np.empty(n + 1)
xx = np.empty(n + 1)
vx = np.empty(n + 1)
ax = np.empty(n + 1)

# Introduzir x0 e v0 nos vetores da posio e velocidade
xx[0] = xx0
vx[0] = vx0
ax[0] = ax0


# Preencher os vetores x, v, a
for i in range(n):    

    ax[i] = (-k * xx[i] - 3 * alpha * xx[i]**2) / m
    vx[i + 1] = vx[i] + ax[i] * dt
    xx[i + 1] = xx[i] + vx[i + 1] * dt
    
    Epot[i] = 0.5 * k * xx[i]**2 + alpha * xx[i]**3

    


# Plot
plt.plot(xx, Epot, label="Epot",  linestyle='-', linewidth=3, color="turquoise")
plt.xlabel("x (t)")
plt.ylabel("t (s)")
plt.legend()         # Legenda s aparece com isto
plt.grid()
plt.show()

plt.plot(t, xx, label="Epot",  linestyle='-', linewidth=3, color="turquoise")
plt.xlabel("x (t)")
plt.ylabel("t (s)")
plt.legend()         # Legenda s aparece com isto
plt.grid()
plt.show()
    
print("Quando o corpo tem menos de 2J de Energia, os seus extremos nao sdao simetricos, sendo que a sua posiao media nao coincide com a posiao de equilibrio. Como a Epot  maior no lado negatico este passa mais tempo nesse lado")
    
```
