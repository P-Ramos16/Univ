```python
import numpy as np
import matplotlib.pyplot as plt

# Gravidade
g = 9.8

# Tempo inicial e final
ti = 0
tf = 1

# Massa
G = 4 * (np.pi ** 2)
M_sol = 1

# Posição inicial
xx0 = -1
xy0 = 0

# Aceleração inicial
ax0 = 0
ay0 = 0

# Velocidade inicial
vx0 = 0
vy0 = -2 * np.pi

# dt incremento do tempo e n numero de intervalos
dt = 0.001
n = int((tf - ti) / dt)

# Vetor tempo (n+1 para garantir que nao falta o ultimo dado (Ex: t[10]))
t = np.linspace(ti, tf, n + 1)

# Vetor velocidade (empty e não zeros para não alterar 
# muito o resultado se faltar analisar um dado)
xx = np.empty(n + 1)
vx = np.empty(n + 1)
ax = np.empty(n + 1)

xy = np.empty(n + 1)
vy = np.empty(n + 1)
ay = np.empty(n + 1)

# Introduzir x0 e v0 nos vetores da posição e velocidade
xx[0] = xx0
vx[0] = vx0
ax[0] = ax0

xy[0] = xy0
vy[0] = vy0
ay[0] = ay0

# Preencher os vetores x, v, a
for i in range(n):
    r = np.sqrt(xx[i] ** 2 + xy[i] ** 2)
    
    ax[i] = (-G * M_sol * xx[i]) / (r ** 3) 
    vx[i + 1] = vx[i] + ax[i] * dt
    xx[i + 1] = xx[i] + vx[i + 1] * dt
    
    ay[i] = (-G * M_sol * xy[i]) / (r ** 3) 
    vy[i + 1] = vy[i] + ay[i] * dt
    xy[i + 1] = xy[i] + vy[i + 1] * dt


# Plot
plt.plot(xx, xy, label="x",  linestyle='-', linewidth=3, color="turquoise")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.legend()         # Legenda só aparece com isto
plt.grid()
```
